### groupby command

```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('movies.csv')
```

```python
data.head(5)
```

```python
genre = data.groupby('Genre')
```

```python
genre.sum()
```

```python
genre.min()
```

```python
genre.max()
```

```python
# find the top three movies based on the genre
data.groupby('Genre')['Gross'].sum().sort_values(ascending=False).head(3)
```

```python
# find the genre with the highest IMDM rating
data.groupby('Genre')['IMDB_Rating'].sum().head(3)
```

```python
# find the genre with the highest IMDM rating
data.groupby('Genre')['IMDB_Rating'].mean().sort_values(ascending=False)
```

```python
# find the director with most popularity
data.groupby('Director')['No_of_Votes'].sum().sort_values(ascending=False).head(3)
```

```python
# Find the highest rated movie of each genre (Wrong answer)
data.groupby('Genre')['IMDB_Rating'].sum().sort_values(ascending=False).head()
```

```python
# Find the highest rated movie of each genre
data.groupby('Genre')['IMDB_Rating'].max()
```

```python
# find the number of movies done by each actor 
data.groupby('Star1').count()['Series_Title'].sort_values(ascending=False)
```

```python
# find the total number of groups
len(data.groupby('Genre'))
```

```python
# find items in each group
data.groupby('Genre').size()
```

```python
# find the starting first data
data.groupby('Genre').size().head()
```

```python
# find the starting last data
data.groupby('Genre').size().tail()
```

```python
# accessing data from particular index
data.groupby('Genre')
```

```python
data.describe()
```

```python
data.groupby('Director').value_counts()
```

```python
# apply aggregation function over multiple columns 
genre = data.groupby('Genre')
```

```python
genre.agg(
    {
        'Released_Year': 'mean',
        'Runtime' : 'max',
        'IMDB_Rating' : 'max',
        'No_of_Votes' : 'max'
    }
)
```
